```python
import pandas as pd

# Load the CSV file
df = pd.read_csv(r'C:\Users\USER\Desktop\netflix_titles.csv')

# Display the first few rows
df.head()
```

```python
# Check number of rows and columns
df.shape

# Show all column names
df.columns

# Info about data types and non-null values
df.info()

# Count of missing values in each column
df.isnull().sum()

# Remove duplicate rows, if any
df.drop_duplicates(inplace=True)

# Confirm shape again after removing duplicates
df.shape
```

```python
# Fill missing values in 'rating' with the mode (most common rating)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)

# Drop rows where 'title' or 'type' is missing (essential fields)
df.dropna(subset=['title', 'type'], inplace=True)

# Fill other missing values with 'Unknown'
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)

# Confirm if nulls are handled
df.isnull().sum()
```

```python
# Standardize column names: lowercase, replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

```python
# Convert 'date_added' from text to datetime (if not unknown)
df['date_added'] = df['date_added'].replace('Unknown', pd.NA)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# Extract month and year separately
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month
```
